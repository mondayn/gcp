```python
# parse certain elements from patient resource of sample fhir from synthea
# easier way to parse: gcp healthcare api

from fhir.resources.bundle import Bundle
from fhir.resources.patient import Patient
from pathlib import Path
import json

def parse_patient(e,file):
    p = Patient.model_construct(**e['resource'])
    return (
        p.name[0]['given'][0],
        p.name[0]['family'],
        p.address[0]['city'],
        p.address[0]['state'],
        Path(file).name
    )

patient_data=[]

files = list(Path(r"C:\Users\hvu7470\Downloads\synthea_sample_data_fhir_latest").glob('*.json'))
file_count=0
for file in files[:]:
    file = Path(file)
    file_count += 1
    with open(file,"r", encoding='utf-8') as f:
        try:
            d = json.load(f)
        except Exception as e:
            print(f'couldnt parse {file.name}')
            
    bundle = Bundle.model_construct(**d)
    patients = [parse_patient(e,file) for e in bundle.entry if e['resource']['resourceType']=='Patient']
    if len(patients)<1: 
        print(f'no patients found for {file.name}')
    else:
        patient_data += patients

print(f'\n{file_count} out of {len(files)} files examined')


df=pd.DataFrame(patient_data,columns=['last','first','city','st','src'])
df.head(5)
df.tail(5)
```

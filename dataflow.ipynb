```python
from google.cloud import bigquery
from gcp_fx import PROJECT_ID
import pandas as pd

table = f'''`{PROJECT_ID}.taxirides.realtime`'''
# bigquery.Client().query(f'drop table {table}')
```

```python
sql = f'''CREATE TABLE {table} (
    ride_id STRING,
    point_idx INTEGER,
    latitude FLOAT64,
    longitude FLOAT64,
    timestamp TIMESTAMP,
    meter_reading FLOAT64,
    meter_increment FLOAT64,
    ride_status STRING,
    passenger_count INTEGER
)
PARTITION BY DATE(timestamp);
'''
bigquery.Client().query(sql)
```

```python
sql = f'select ride_id, ride_status, passenger_count, count(*) over() as ttl_records from {table}'
bigquery.Client().query(sql).to_dataframe()
```

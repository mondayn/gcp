```python
tkn = !gcloud auth print-identity-token

# posts a url to the cloud function endpoint
# returns ai summary
def get_ai_summary(url):
    import requests
    headers = {
        "Authorization": "bearer "+tkn[0]
        ,"Content-Type": "application/json"
    }
    response = requests.post(
        "https://us-east4-hca-hvu7470-202510281151-sbx.cloudfunctions.net/summarize2"
        , headers=headers, json={"url":url}
    )
    print(
        'response ',response.status_code,'\n'
        'length=',len(response.text)
    )
    return response.text
```
